```python
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('../datasets/sample/sample_labels.csv')
```

```python
df.head()
```

```python
df.columns = ['image_index', 'finding_labels', 'follow_up_number', 'patient_id',
       'patient_age', 'patient_gender', 'view_position', 'original_image_width',
       'original_image_height', 'original_image_pixel_spacing_x',
       'original_image_pixel_spacing_y']
```

```python
df['finding_labels'] = df['finding_labels'].str.lower()
```

```python
df.head()
```

```python
classes = [
    'atelectasis', 
    'consolidation', 
    'infiltration', 
    'pneumothorax', 
    'edema', 
    'emphysema', 
    'fibrosis', 
    'effusion', 
    'pneumonia', 
    'pleural_thickening', 
    'cardiomegaly', 
    'nodule', 
    'mass', 
    'hernia'
]
```

```python
for col in classes:
    df[col] = df['finding_labels'].apply(lambda x: 1 if col in x else 0)
```

```python
df.drop('finding_labels', axis=1, inplace=True)
```

```python
df.head()
```

```python
df['patient_age'] = df['patient_age'].apply(lambda x: int(x[:-1]) if x.endswith('Y') else int(x[:-1]) / 12 if x.endswith('M') else x)
```

```python
df['patient_age'] = df['patient_age'].apply(lambda x: 0 if str(x).endswith('D') else x)
```

```python
df['patient_age'] = df['patient_age'].astype(int)
```

```python
df['patient_age'].sample(10)
```

```python
df[df['patient_age'] <= 2]
```

```python
df['patient_gender_M'] = df['patient_gender'].apply(lambda x: 1 if x == 'M' else 0)
df['patient_gender_F'] = df['patient_gender'].apply(lambda x: 1 if x == 'F' else 0)
```

```python
df.drop('patient_gender', axis=1, inplace=True)
```

```python
df['view_position_AP'] = df['view_position'].apply(lambda x: 1 if x == 'AP' else 0)
df['view_position_PA'] = df['view_position'].apply(lambda x: 1 if x == 'PA' else 0)
```

```python
df.drop('view_position', axis=1, inplace=True)
```

```python
df.head()
```

```python
df.to_csv('../datasets/sample/sample_labels_clean.csv', index=False)
```
